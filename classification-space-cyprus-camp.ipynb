### Introduction

*Year 2222*. The modern cosmologist Arkadiya knows for sure that theoretically there should be a balance of *ordinary* and *dwarf* galaxies in the Universe. 
However, the problem of the shortage of dwarf galaxies, known since the 20th century, is that the number of observed dwarf galaxies is many times less than predicted by theory.

Arkadiya has a collected dataset of **1000 observed galaxies**, which are described by **13 numerical features**; the class label is known for them (0 - dwarf, 1 - ordinary).

Help Arkadiya train a **binary classifier** that will work as accurately as possible on real data — all galaxies.

*P.S. The real data is a thousand times larger and confirms the theoretical distribution of ordinary and dwarf galaxies.*

- The `test.csv` file contains data on unobserved galaxies that you have to label
- The `sample_submission.csv` file is an example of a correct submission

#### What should be done
Upload the `submission.csv` file to the [Kaggle Competition](https://www.kaggle.com/competitions/cyprus-ai-camp-binary-classification-in-space), each line of which contains the class (in the `target` column) and the `Id` of the galaxy, separated by a comma.

Your score on this problem will be calculated as:

$$\min(100, (\text{accuracy} - 0.55)*294),$$

where `accuracy` is the proportion of correctly classified galaxies, according to your predictions.

```python
folder = "/kaggle/input/cyprus-ai-camp-binary-classification-in-space/"
data = pd.read_csv(folder + 'train.csv')
data_test = pd.read_csv(folder + 'test.csv')
data.head(5)
```

```python
X = data.drop('target', axis=1)
y = data['target']
```

### XGBoost

```python
from xgboost import XGBClassifier
from catboost import CatBoostClassifier, Pool


scale_pos_weight = (y == 0).sum() / (y == 1).sum()

model = CatBoostClassifier(scale_pos_weight=scale_pos_weight, verbose=0)

model.fit(X, y)
```

```python
import string
import random
import numpy as np


test = pd.read_csv('/kaggle/input/cyprus-ai-camp-binary-classification-in-space/test.csv')

#preds = pipe.predict(test)


proba = model.predict_proba(test)[:, 1]

threshold = np.quantile(proba, 0.5)

preds = (proba >= threshold).astype(int) 

submit = pd.DataFrame()
submit['target'] = preds
submit['Id'] = range(len(test))

all_chars = string.ascii_letters + string.digits
random_chars = ''.join(random.sample(all_chars, 6))
submit.to_csv(f'submission{random_chars}.csv', index=False)

submit['target'].value_counts()
```
